```python
# 載入基本套件
import pandas as pd
import numpy as np

# 讀取訓練與測試資料
data_path = '../data/'
df_train = pd.read_csv(data_path + 'titanic_train.csv')
df_test = pd.read_csv(data_path + 'titanic_test.csv')
df_train
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
# 重組資料成為訓練 / 預測用格式
train_Y = df_train['Survived']
ids = df_test['PassengerId']
df_train.drop(['PassengerId', 'Survived'] , axis=1, inplace = True)
df_test.drop(['PassengerId'] , axis=1, inplace = True)
df = pd.concat([df_train,df_test])
df.reset_index(inplace = True, drop = True)
df
```

```python
# 秀出資料欄位的類型與數量
df.dtypes.value_counts()
```

```python
#確定只有 int64, float64, object 三種類型後, 分別將欄位名稱存於三個 list 中
int_features = df.select_dtypes('int64').columns.tolist()
float_features = df.select_dtypes('float64').columns.tolist()
object_features = df.select_dtypes('object').columns.tolist()
```
