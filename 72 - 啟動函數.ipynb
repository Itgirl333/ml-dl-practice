# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''

def ReLU(x):
    return x*(x>0)

def dReLU(x):
    return 1*(x>0)

x = np.linspace(-10,10,100)

plt.plot(x, ReLU(x), label='ReLU(x)')
plt.plot(x, dReLU(x), label='dReLU(x)')
plt.legend()
plt.show()
```
