### 用 digits (dataset.load_digits()) 資料嘗試跑 t-SNE

```python
from sklearn import datasets
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
digits = datasets.load_digits(n_class=4)
X = digits.data
y = digits.target
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components = 2, random_state = 0, init = 'pca', learning_rate = 200., early_exaggeration = 12.)
X_tsne = tsne.fit_transform(X)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c = y)
for name, label in [('class0', 0), ('class1', 1),('class2', 2),('class3', 3)]:
    plt.text(X_tsne[y == label, 0].mean(),
            X_tsne[y == label, 1].mean(), name,
            horizontalalignment = 'center',
            bbox = dict(alpha = .5, edgecolor = 'w', facecolor = 'w'))
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components = 3, random_state = 0, init = 'pca', learning_rate = 200., early_exaggeration = 12.)
X_tsne = tsne.fit_transform(X)

fig = plt.figure(figsize = (4, 3))
ax = Axes3D(fig, rect = [0, 0, .95, 1], elev = 48, azim = 134)
ax.scatter(X_tsne[:, 0], X_tsne[:, 1], X_tsne[:, 2], c = y, edgecolor = 'k')  # c = y時，代表 Ground Truth
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
for name, label in [('0', 0), ('1', 1), ('2', 2), ('3', 3)]:
    ax.text3D(X_tsne[y == label, 0].mean(),
              X_tsne[y == label, 1].mean() + 1.5,
              X_tsne[y == label, 2].mean(), name,
              horizontalalignment = 'center',
              bbox = dict(alpha = .5, edgecolor = 'w', facecolor = 'w'))
ax.set_title('TSNE n_components = 3')
```
