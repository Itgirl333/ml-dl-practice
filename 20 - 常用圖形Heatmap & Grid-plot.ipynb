```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
matrix = np.random.uniform(-1.0, 1.0, size = (10, 10))

plt.figure(figsize=(10,10)) 
sns.heatmap(np.corrcoef(matrix), cmap = plt.cm.RdYlBu_r, annot = True)
plt.show()
```

```python
nrow = 1000
ncol = 3

matrix = np.random.uniform(-1.0, 1.0, size = (nrow, ncol))

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size = nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()
plot_data.rename(columns = {0 : 'col1', 1 : 'col2', 2 : 'col3'}, inplace = True)
# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False, 
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])

grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```

```python
sns.kdeplot(plot_data.loc[plot_data['index'] == 0, 'col3'])
sns.kdeplot(plot_data.loc[plot_data['index'] == 1, 'col3'])
sns.kdeplot(plot_data.loc[plot_data['index'] == 2, 'col3'])
```

```python
sns.kdeplot(plot_data['col3'], hue=plot_data['index'])
```

```python
nrow = 1000
ncol = 3

matrix = np.random.randn(nrow, ncol)

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size = nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()
plot_data.rename(columns = {0 : 'col1', 1 : 'col2', 2 : 'col3'}, inplace = True)
# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False,
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])

grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```
